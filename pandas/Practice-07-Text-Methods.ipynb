## Section 3 — Cleaning pipeline

### Exercise 3.1
Build a full cleaning pipeline for `raw_titles` in a single chained expression:
1. Strip whitespace
2. Replace `;` with space
3. Replace `_` with space
4. Apply `.title()` capitalization

Save the result as `clean_titles`.

````python
# Your code here
clean_titles = raw_titles.str.strip().str.replace(';',' ').str.replace('_',' ').str.title()
display(raw_titles)
display(clean_titles)
````

### Exercise 3.2
Replicate the same pipeline using a custom `cleanup()` function + `np.vectorize()`.
Time both approaches with `%%timeit` (one cell each) and note which is faster.
*(This is a real-world skill — vectorize wins on large datasets)*

````python
def cleanup(title):
    title = title.strip()
    title = title.replace(';', ' ')
    title = title.replace('_', ' ')
    title = title.title()
    return title
````

````python
# Time the .str chain approach
%%timeit
print(raw_titles.str.strip().str.replace(';',' ').str.replace('_',' ').str.title().values)
````

````python
# Time the np.vectorize approach
# %%timeit
print(np.vectorize(cleanup)(raw_titles).tolist())
````

You receive this raw Series of game entries from a scraped webpage:
```python
scraped = pd.Series([
    "  grand theft AUTO v  | action | 2013 | take-two  ",

```

**Step 1:** Split each entry on `'|'` with `expand=True` into 4 columns: `title`, `genre`, `year`, `publisher`.

**Step 2:** Strip whitespace from all 4 columns.

**Step 3:** Apply `.str.title()` to `title` and `publisher`. Apply `.str.lower()` to `genre`.

**Step 4:** Convert `year` to integer.

**Step 5:** Add a column `'title_length'` = number of characters in the cleaned title.

**Step 6:** Filter and show only games where `genre` contains `'action'`.

````python
scraped = pd.Series([
    '  grand theft AUTO v  | action | 2013 | take-two  ',
    'HALO 3|shooter|2007|microsoft',
    ' minecraft | sandbox | 2013 | mojang ',
    'FIFA 14|sports|2013|EA Sports',
    '  GOD OF WAR | action-adventure | 2005 | sony  '
])
# Step 1
clean = scraped.str.split('|', expand = True)
clean.columns=['title', 'genre','year','publisher']
clean
````

````python
# Step 2
for col in clean:
    clean[col] = clean[col].str.strip()
clean
````

````python
# Step 3
clean['title'] = clean['title'].str.title()
clean['publisher'] = clean['publisher'].str.title()
clean['genre'] = clean['genre'].str.lower()
clean
````

````python
# Step 4
clean['year'] = clean['year'].astype(int)
clean
````

````python
# Step 5
clean['title_lenght'] = clean['title'].str.len()
clean

````

````python
# Step 6
clean[clean['genre'].str.contains('action')]
````
