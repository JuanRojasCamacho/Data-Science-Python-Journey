## 🔥 Challenge — Exercise 3.4
You have monthly revenue (in $M) for two games:

```python
game_a = {'Jan': 1.2, 'Feb': 0.9, 'Mar': 1.5, 'Apr': 1.8}
game_b = {'Jan': 0.8, 'Feb': 1.1, 'Mar': 1.3, 'Apr': 2.1, 'May': 1.7}
```

1. Convert both to Series.
2. Calculate `total_revenue` by adding them with `fill_value=0`.
3. Multiply the total by 1,000,000 to convert to full dollars.
4. Print the month where total revenue was highest. *(Hint: use `.idxmax()`)*


````python
game_a = {'Jan': 1.2, 'Feb': 0.9, 'Mar': 1.5, 'Apr': 1.8}
game_b = {'Jan': 0.8, 'Feb': 1.1, 'Mar': 1.3, 'Apr': 2.1, 'May': 1.7}
# Your code here
a_serie = pd.Series(game_a)
b_serie = pd.Series(game_b)

total_revenue = a_serie.add(b_serie, fill_value = 0) * 1000000
print(f"Highest revenue: {total_revenue.idxmax()}")

````
