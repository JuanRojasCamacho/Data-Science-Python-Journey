## 🔥 Final Challenge

Using `df` (with `date` as index), answer these questions step by step:

**Step 1:** Resample by month (`'ME'`) and sum `units_sold`. Save as `monthly`.

**Step 2:** Find the single **best month** (highest units) and the **worst month** across all data.
Print both the date and the value.

**Step 3:** Reset the index of `monthly`. Add a `'year'` and `'month_name'` column using `.dt`.

**Step 4:** Group by `'month_name'` and compute mean `units_sold`. 
Which calendar month is consistently the best performer?

**Step 5:** Resample the original `df` by year and compute total `revenue_usd`.
Calculate the **year-over-year growth rate** (%) between each year.
*(Hint: use `.pct_change() * 100`)*

```python
# Step 1
monthly = df['units_sold'].resample('ME').sum()
monthly
```

```python
# Step 2
best = monthly.idxmax().strftime('%B %Y')
worst = monthly.idxmin().strftime('%B %Y')
print(f"Best month: {best}")
print(f"Worst month: {worst}")
```

```python
# Step 3
monthly = monthly.reset_index()
monthly['year'] = monthly['date'].dt.year
monthly['month_name'] = monthly['date'].dt.month_name()
monthly.head()
```

```python
# Step 4
monthly.groupby('month_name').mean()['units_sold'].idxmax()
```

```python
# Step 5
yearly_rev = df['revenue_usd'].resample('YE').sum()
yoy = (yearly_rev.pct_change() * 100).round(1)
print("\nYear-over-year revenue growth (%):")
print(yoy)
```
