### Exercise 1.1
Write a function called `sales_tier` that takes a `Global_Sales` value and returns:
- `'Blockbuster'` if sales >= 30M
- `'Hit'` if sales >= 10M
- `'Mid'` if sales >= 5M
- `'Niche'` otherwise

Apply it to the `Global_Sales` column and store the result in a new column called `'Tier'`.
Display the first 10 rows showing `Title`, `Global_Sales`, and `Tier`.


````python
# Your code here
def sales_tier(global_sales):
    if global_sales >= 30:
        return 'Blockbuster'
    elif global_sales >=10:
        return 'Hit'
    elif global_sales >= 5:
        return 'Mid'
    else:
        return 'Niche'
df['Tier'] = df['Global_Sales'].apply(sales_tier)
df.head()
````

## Section 3 — `.apply()` across multiple columns

### Exercise 3.1
Write a function called `market_focus` that takes `na_sales` and `eu_sales` and returns:
- `'NA Dominant'` if NA_Sales > EU_Sales * 1.5
- `'EU Dominant'` if EU_Sales > NA_Sales * 1.5
- `'Balanced'` otherwise

Apply it using a lambda across `['NA_Sales', 'EU_Sales']` with `axis=1`.
Store in a new column `'Market_Focus'`.


````python
# Your code here
def market_focus(na_sales, eu_sales):
    if na_sales > eu_sales*1.5:
        return 'NA Dominant'
    elif eu_sales > na_sales*1.5:
        return 'EU Dominant'
    else:
        return 'Balanced'


df['Market_Focus'] = df[['NA_Sales', 'EU_Sales']].apply(
    lambda row: market_focus(row['NA_Sales'], row['EU_Sales']), axis=1
)
df.head()
````

### Exercise 3.2 — `np.vectorize` (faster alternative)
Rewrite the same `market_focus` logic but this time apply it using `np.vectorize`
instead of `.apply()`. Store the result in `'Market_Focus_v2'`.

Then confirm both columns are identical using `==` and `.all()`.


````python
# Your code here
df['Market_Focus_V2'] = np.vectorize(market_focus)(df['NA_Sales'], df['EU_Sales'])

df['Comparison'] = (df['Market_Focus_V2'] == df['Market_Focus']).all
df.head()
````
