## Section 1 — Single Conditions

### Exercise 1.1
Filter all games where `Global_Sales` is greater than **20 million**.

```python
# Your code herep
display(Markdown("### Sales over 20 million"))
display(df[df['Global_Sales'] > 20])
```

### Exercise 1.2
Filter all games published by `'Nintendo'`.

```python
# Your code here
display(Markdown("### Nintendo Games:"))
display(df[df['Publisher'] == 'Nintendo'])
```

### Exercise 1.3
Filter all games where the `Genre` is **NOT** `'Sports'`. Use the `~` operator.

```python
# Your code here
display(Markdown("### No sports games:"))
display(df[df['Genre'] != 'Sports'])
```

## Section 2 — Multiple Conditions

### Exercise 2.1 — AND condition
Filter games that are:
- Genre is `'Shooter'` **AND**
- `NA_Sales` is greater than 7

```python
# Your code here
display(Markdown("### Shooter and Sales over 7 Games:"))
display(df[ (df['Genre'] == 'Shooter') & (df['NA_Sales'] > 7) ])
```

### Exercise 2.2 — OR condition
Filter games released in `1985` **OR** `1989`.

```python
# Your code here
display(Markdown("### Games releades on 1985 or 1989:"))
display(df[df['Year'].isin([1985,1989])])
```

### Exercise 2.3 — Combining AND + NOT
Filter games where:
- `Global_Sales` > 10 million **AND**
- Publisher is **NOT** `'Nintendo'`

```python
# Your code here
display(Markdown("### Sales over 10 and publisher NOT nintendo"))
display(df[(df['Global_Sales'] > 10) & (df['Publisher'] != 'Nintendo')])
```

### Exercise 2.4 — Three conditions
Filter games where:
- `Genre` is `'Action'` **AND**
- `EU_Sales` > 5 **AND**
- `Critic_Score` > 90

How many results do you get?

```python
# Your code here
display(Markdown("### Action, EU Sales over 5 and Critic Score over 90"))
conditions = (df['Genre'] == 'Action') & (df['EU_Sales'] > 5) & (df['Critic_Score'] > 90)
display(df[conditions])
```

## Section 3 — `.isin()`

### Exercise 3.1
Filter games where the `Platform` is in this list: `['PS2', 'PS3', 'X360']`.

```python
# Your code here
list = ['PS2', 'PS3', 'X360']
display(df[df['Platform'].isin(list)])
```

### Exercise 3.2 — NOT isin()
Filter games where the `Genre` is **NOT** in `['Sports', 'Racing', 'Misc']`.

```python
# Your code here
list2 = ['Sports', 'Racing', 'Misc']
display(df[~df['Genre'].isin(list2)])
```

## 🔥 Challenge — Exercise 3.3
You're a market analyst. Answer these questions using conditional filtering:

1. **Question A:** Which games sold more than 5M in Europe (`EU_Sales > 5`) AND were released after 2005? Show only `Title`, `Genre`, `EU_Sales`, and `Year`.

2. **Question B:** How many unique genres exist among games that sold more than 15M globally?
*(Hint: filter first, then use `.nunique()` or `.unique()` on the Genre column)*

3. **Question C:** Filter all Nintendo games on handheld platforms (`['GB', 'DS', 'GBA']`). What is the total combined `Global_Sales` for this group?
*(Hint: filter, then use `.sum()` on the Global_Sales column)*


```python
# Question A
display(Markdown("### EU SALES over 5 AND released afte 2005"))
conditions1 = (df['EU_Sales'] > 5) & (df['Year'] > 2005)
display(df.loc[conditions1, ['Title', 'Genre', 'EU_Sales', 'Year']])
```

```python
# Question B
display(Markdown("### Unique Genres with sells more than 15M"))
filter = df[df['Global_Sales'] > 15]
display(filter['Genre'].nunique())
```

```python
# Question C
display(Markdown("### Nintendo on Handheld platforms"))
conditions2 = (df['Publisher'] == 'Nintendo') & (df['Platform'].isin(['GB', 'DS', 'GBA']))
display(df[conditions2]['Global_Sales'].sum())
```
