**Step 1:** Count total missing values per column.

**Step 2:** Drop the row that is **completely empty** (all NaN). *(Hint: `thresh=1` keeps rows with at least 1 non-null)*

**Step 3:** Drop columns where more than 40% of values are missing.

**Step 4:** Fill missing `Critic_Score` with the mean. Fill missing `Publisher` with `'Unknown'`.

**Step 5:** How many rows still have **any** NaN after steps 2–4? Which columns still have gaps?

````python
# Step 1
print(messy.isnull().sum())
````

````python
# Step 2
messy.dropna(thresh=1)
messy.shape
````

````python
# Step 3
percent = int(len(messy.columns)*0.6)
messy.dropna(thresh = percent ,axis = 1)
print(f"Columns remaining: {list(messy.columns)}")
````

````python
# Step 4
messy['Critic_Score'] = messy['Critic_Score'].fillna(messy['Critic_Score'].mean())
messy['Publisher'] = messy['Publisher'].fillna('Unknown')
messy
````

````python
# Step 5
still = messy.isnull().sum()
print(f"Rows with any NaN: {messy.isnull().any(axis=1).sum()}")
print("Columns still with gaps:")
print(still[still > 0])
````
