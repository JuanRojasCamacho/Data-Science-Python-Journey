## Section 1 — Basic `groupby()`

### Exercise 1.1
Group by `Genre` and calculate the **mean** of all numeric columns.
Which genre has the highest average `Global_Sales`?


```python
# Your code here
group_genre = df.groupby('Genre').mean(numeric_only = True)
print(f"Highest Average : {group_genre['Global_Sales'].idxmax()}")

```

### Exercise 1.2
Group by `Publisher` and get the **sum** of `NA_Sales`, `EU_Sales`, `JP_Sales`, and `Global_Sales`.
Sort by `Global_Sales` descending. Who dominates?

```python
# Your code here
group_publisher = df.groupby('Publisher').sum()[['NA_Sales','EU_Sales','JP_Sales','Global_Sales']].sort_values('Global_Sales', ascending = False)
print(f"Publisher that dominates: {group_publisher['Global_Sales'].idxmax()}")
```

### Exercise 1.3
Group by `Platform` and use `.describe()` on `Global_Sales`.
Transpose the result. Which platform has the highest `max`?

```python
# Your code here
group_plat = df.groupby('Platform').describe()['Global_Sales'].transpose()
print(f"Platform with highest max: {group_plat.loc['max'].idxmax()}")
```

## Section 4 — `agg()` for custom aggregations

### Exercise 4.1
On the full `df`, use `.agg()` to compute `['mean', 'median', 'std']` for the entire DataFrame.
Display only the columns `Global_Sales`, `NA_Sales`, `EU_Sales`.

```python
# Your code here
df[['Global_Sales','NA_Sales','EU_Sales']].agg(['mean','median','std'])
```

### Exercise 4.2
Use `.agg()` with a dictionary to compute:
- `Global_Sales`: mean and max
- `Critic_Score`: median and min
- `NA_Sales`: sum and mean

```python
# Your code here
df.agg({
    'Global_Sales':['mean','max'],
    'Critic_Score':['median','min'],
    'NA_Sales':['sum','mean']
    })
```

### Exercise 4.3 — `groupby` + `agg()`
Group by `Genre` and aggregate:
- `Global_Sales`: mean and sum
- `Critic_Score`: median (ignore NaN)
- `NA_Sales`: mean

Sort the result by `('Global_Sales', 'mean')` descending.

```python
# Your code here
result = df.groupby('Genre').agg({
    'Global_Sales': ['mean', 'sum'],
    'Critic_Score' : ['median'],
    'NA_Sales': ['mean']   
})
result
```

## Section 5 — Filter before groupby (good practice)

### Exercise 5.1
The course warns: *it's much easier to filter before groupby than after.*
Group only `'Action'` and `'Shooter'` games by `Publisher` and compute total `Global_Sales`.
Filter **before** the groupby call.

```python
# Your code here
df[df['Genre'].isin(['Action','Shooter'])].groupby('Publisher')['Global_Sales'].sum()
```

### Exercise 5.2
Among games released **after 2000**, group by `Genre` and get the mean `Critic_Score`.
Sort descending. Drop NaN groups.

```python
# Your code here
df[df['Year'] > 2000].groupby('Genre')['Critic_Score'].mean().sort_values(ascending = False)
```
