## Section 2 — `pd.merge()` join types

### Exercise 2.1 — Inner join
Merge `games` and `sales` on `game_id` using an **inner join**.
How many rows does the result have, and which games were dropped?

````python
# Your code here
display(games)
display(sales)
display(pd.merge(games,sales,how='inner',on='game_id'))

#Game number 6 and 9 gets dropped because is not shared on both tables
````

Merge `games` and `sales` on `game_id` using a **left join**.

````python
# Your code here
display(games)
display(sales)
display(pd.merge(games,sales,how='left',on='game_id'))

#Only number 9 gets dropped because all the games from table games are included
````

### Exercise 3.1
Chain two merges:

## 🔥 Final Challenge

Build the most complete dataset possible from all four tables.

**Step 1:** Concat `publishers_a` and `publishers_b` into `publishers`. Reset index.

**Step 2:** Inner join `games` with `sales` on `game_id`. Save as `gs`.

**Step 3:** Left join `gs` with `scores` on `game_id`. Save as `gss`.

**Step 4:** Left join `gss` with `publishers` on `title`. Save as `full`.

**Step 5:** Print the shape of `full`. How many games are missing a publisher?

**Step 6:** Fill missing `publisher` values with `'Unknown'`.

**Step 7:** Show the top 5 games by `global_sales` with columns: `title`, `publisher`, `genre`, `global_sales`, `critic_score`.

````python
# Step 1
publishers = pd.concat([publishers_a, publishers_b]).reset_index(drop=True)
display(publishers)
````

````python
# Step 2
gs = pd.merge(games,sales, how = 'inner', on='game_id')
display(gs)
````

````python
# Step 3
gss = pd.merge(gs,scores,how = 'left', on='game_id')
````

````python
# Step 4
full = pd.merge(gss, publishers, how='left', on='title')
display(full)
````

````python
# Step 5
print("Shape:", full.shape)
print("Missing publisher:", full['publisher'].isnull().sum())
````

````python
# Step 6
full['publisher'] = full['publisher'].fillna('Unknown')
display(full)
````

````python
# Step 7
full[['title','publisher','genre','global_sales','critic_score']].nlargest(5,'global_sales')
````
